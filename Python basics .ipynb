```python
def factorial(n):
    result = 1
    if n == 1 or n == 0:
        return result
    elif n<0:
        return "not defined, use positive numbers";
    else:
        for i in range(2,n+1):
            result *= i
    return result    
```

```python
factorial(5)
```
